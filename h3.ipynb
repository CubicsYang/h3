```python
import geopandas
import h3
import shapely.geometry
import matplotlib.pyplot as plt
geoJson = {'type': 'Polygon',
 'coordinates': [[[32.2764916, 118.5172165], [31.86323241, 118.5172165], [31.86323241, 118.91648798], [32.2764916, 118.91648798]]] }
hexagons = list(h3.polyfill(geoJson, 7))
gdf = geopandas.GeoDataFrame(geometry=[shapely.geometry.Polygon(h3.h3_to_geo_boundary(h,geo_json=False)) for h in hexagons])
# plot the GeoDataFrame
fig, ax = plt.subplots(figsize=(10, 10))
gdf.boundary.plot(ax=ax)
gdf.plot(ax=ax, facecolor='lightgrey')
# set title
plt.title('Polygon filled with H3 hexagons')
plt.show()
```
